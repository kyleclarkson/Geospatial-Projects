```python
# Import module
import osmnx as ox

# Area of interest (for new area, see nominatim openstreetmap and test queries.)
place_name = "kamppi, Helsinki, Finland"

# Get OSM street network
graph = ox.graph_from_place(place_name)
```

```python
type(graph)
```

```python
# Plot graph
ox.plot_graph(graph)
```

```python
# Get geodataframe 

area = ox.gdf_from_place(place_name)
```

```python
type(area)
```

```python
area
```

```python
#plot area using coordinates (using Matplotlib)

area.plot()
```

```python
# Building footprints

buildings = ox.footprints_from_place(place_name)
```

```python
# Points of interest
restaurants = ox.pois_from_place(place_name, {'amenity':True})
```

```python
# Get graph as vertices and edges
nodes, edges = ox.graph_to_gdfs(graph)
```

```python
nodes.head()
```

```python
edges.head()
```

```python
# Plotting the data

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12,8))

# Plot footprint
area.plot(ax=ax, facecolor="black")

# Plot edges
edges.plot(ax=ax, linewidth=1)

# Plot buildings
buildings.plot(ax=ax, facecolor="khaki", alpha=0.7)

# Plot restaurants 
restaurants.plot(ax=ax, color="red", markersize=10)
```
